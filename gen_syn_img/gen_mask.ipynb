```python
import random
import numpy as np
import cv2
import json
import os
import matplotlib.pyplot as plt
```

Create mask for one image

```python
# imgpath = '../dataset/inpaint/images/test/00a36f96e31731c4.jpg'
imgpath = 'after_augmentation.jpg'

image = cv2.imread(imgpath, cv2.IMREAD_GRAYSCALE)

threshold_value = np.percentile(image, 100 - 30)  # Top 30% of intensity values
mask = np.where(image >= threshold_value, 255, 0).astype(np.uint8)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')                                                                                                                                                       

plt.subplot(1, 2, 2)
plt.imshow(mask, cmap='gray')
plt.title('Mask')


plt.tight_layout()
plt.show()

cv2.imwrite('mask.jpg', mask)
```

```python
def gen_mask(img_dir, mask_dir):

	for filename in os.listdir(img_dir):
		file_path = os.path.join(img_dir, filename)
		if not os.path.exists(mask_dir):
			os.makedirs(mask_dir)
		
		if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.gif')):
			img_file_path = os.path.join(img_dir, filename)
			mask_file_path = os.path.join(mask_dir, filename)
			if not os.path.exists(mask_file_path):
				image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
				threshold_value = np.percentile(image, 100 - 30)  # Top 30% of intensity values
				mask = np.where(image >= threshold_value, 255, 0).astype(np.uint8) 
				cv2.imwrite(mask_file_path, mask)
				print(f"Created and saved mask: {filename}")
```

```python
# Define your source and destination directories
img_directory   = '../dataset/inpaint/images/train'
mask_directory  = '../dataset/inpaint/masks/train'

gen_mask(img_directory, mask_directory)
```
